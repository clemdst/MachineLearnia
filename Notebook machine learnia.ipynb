$for$ : range($x$) affiche les nombre de $0$ à $x-1$. Range prend en réalité trois arguments, dont deux par défaut : début de la séquence (0 par défaut), longueur (variable) et pas (1 par défaut).

Ainsi, $range(5,10,2)$ donne $5, 7, 9$. On pourrait aussi écrire $range(10,-10,-3)$ pour obtenir $10, 7, 4, ...$

$while$ : mélange des deux structures précédentes.

On peut aussi retourner des booléens avec des fonctions, par exemple def ..... return $x > 0$ renvoie la valeur "True" ou "False"

```python
def fibo(n):
    x=0
    y=1
    liste = [0]
    while y < n:
        print(x)
        x, y = y, x + y
        liste.append(x)
    return liste

# !!! on doit absolument faire la mise à jour en simultané, donc écrire les égalités sur une seule ligne !!!
    
```

## Dictionnaires

Structure de données non-ordonné : seulement système **d'association** clé-valeur et non **d'affectation**

!! jamais deux fois la même clé !! mais même valeur est possible

On peut créer un dictionnaire de dictionnaires : *nested dictionnaire*

C'est très utilisé pour stocker des paramètres (notamment les réseaux de neuronnes), et ensuite l'utiliser pour le modifier

Pour afficher les valeurs : $dic.values()$, $dic.keys()$

Pour sa taille : $len(dic)$

Pour ajouter une association : *dic['nouvelle_clé'] = nouvelle_valeur*

La méthode *get* permet de chercher un élément dans un dictionnaire si l'on n'est pas sûr de son existence, avec *dic.get('clé')*

A partir d'une liste donnée *liste*, on peut la rajouter au dictionnaire avec *dic.fromkeys(liste)*. Pour lui attribuer une valeur par défaut, on fait *dic.fromkeys(liste, 'default')*, ce qui attribue la valeur 'default' à toutes les nouvelles clés.

La méthode *pop* permet de supprimer une clé du dictionnaire et de stocker la valeur qui lui était attribuée. Par exemple, *dic.pop('élément)* supprime *"élément"* du dictionnaire et renvoie *valeur_élément*

```python
dico = {
    "chien":1,
    "chat":2,
    "hibou":3
}

# Pour récupérer les clés :

for i in dico :
    print(i)


# Pour récupérer les valeurs : 

for i in dico.values():
    print(i)

# Pour récupérer la paire des deux :

for i,k in dico.items():
    print(i,k)
```

```python
classeur = {
    "positif":[],
    "négatif":[],
    "nul":[]
}


def trier(classeur, nombres):
    for i in nombres:
        if i > 0 :
            classeur['positif'].append(i)
        elif i < 0 :
            classeur['négatif'].append(i)
        else : 
            classeur['nul'].append(i)
    return classeur
```

## Broadcasting

Le système d'itération par calcul avec boulces for etc est déjà implémenté dans numpy, on peut directement faire A+B pour sommer des matrices et pas la somme élément par élément.

Le broadcasting consiste à "étendre" les dimensions d'un tableau ; par exemple sommer une matrice et un nombre, c'est comme si on sommait la matrice pleine du nombre. Il fonctionne ssi les dimensiosn sont égales, ou si l'une est égale à 1 et les autres égales.

```python
# Exemple

np.random.seed(0)
A = np.random.randint(0,10, (4,1))
B = np.ones((1,4))

print(A,B,A+B) # donne une matrice 4x4 alors qu'on somme deux vecteurs ! on dédouble les lignes et colonnes qui manquent.
```

```python
# !! aux dimensions incomplètes de forme (3,). Le 3 sera considéré comme étant la seconde dimension ! 
# Si on somme deux tableaux de dimensions respectives (100,1) et (100,), on obtient un tableau de dimension (100,100) !

np.random.seed(0)
A = np.random.randint(0,10, (4,))
B = np.ones((1,4))

print(A,B,A+B)
```

### Optimize

```python
# La fonction curve_fit nous permet de trouver une droite qui va bien sur un modèle à définir

# Exemple avec :
x = np.linspace(0,2,100)
y = 1/3*x**3 - 3/5*x**2 + 2 + np.random.randn(x.shape[0])/20

def f(x,a,b,c,d):
    return a*x**3 + b*x**2 + c*x + d

from scipy import optimize

optimize.curve_fit(f,x,y) # donne deux tableaux : les paramètres optimaux et leur covariance

param, cov = optimize.curve_fit(f,x,y)
```

### Minimize

```python
def f(x):
    return x**2 + 15*np.sin(x)

x = np.linspace(-10,10,100)
plt.plot(x,f(x))
```

### Traitement du signal

```python
# Signal processing

x = np.linspace(0,20,100) # x.shape=100
y = x + 4*np.sin(x) + np.random.randn(x.shape[0]) # y est de taille 100, on ajoute un bruit à chacune de ses coordonnées

plt.plot(x,y)
```

```python
# Transformée de Fourier permet d'analyser les fréquences dans un signal périodique, de décomposer un signal en différentes fréquences qui le composent

x = np.linspace(0,30,1000)
y = 3*np.sin(x) + 2*np.sin(5*x) + np.sin(10*x)
plt.plot(x,y)
```

```python
from scipy import fftpack

fourier = fftpack.fft(y) # décomposition
power = np.abs(fourier) # pas prendre les valeurs négatives
freq = fftpack.fftfreq(y.size) # signaux

plt.plot(np.abs(freq), power) # on obtient un graphe qui donne les principales fréqauences
```

```python
# Très utile pour nettoyer un signal bruité ! 

x = np.linspace(0,30,1000)
y = 3*np.sin(x) + 2*np.sin(5*x) + np.sin(10*x) + np.random.random(x.shape[0])*10
plt.plot(x,y)
```

```python

fourier = fftpack.fft(y) # décomposition
power = np.abs(fourier) # pas prendre les valeurs négatives
freq = fftpack.fftfreq(y.size) # signaux

plt.plot(np.abs(freq), power) # le spectre est très bruyant ! il faut le nettoyer
```

```python
# on le filtre : 

fourier[power<400] = 0

plt.plot(np.abs(freq), np.abs(fourier)) # le graphe est propre
```

```python
# on peut alors appliquer la transformation inverse de Fourier

filtré = fftpack.ifft(fourier)

plt.plot(x,y)
plt.plot(x,filtré, c='r', lw=3)
```

## Entraînement du modèle

```python
# Diviser les données en train et test pour évaluer le modèle

from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

data = load_iris()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)

from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
model.fit(X_train, y_train) 

# puis tester avec les données de X_test et y_test  : model.score(X_test, y_test)
```

```python
# On doit ensuite régler les hyperparamètres du modèle, en évitant l'overfitting. Pour cela, on crée un troisème jeu de données,
# appelé validation set, qui permet d'hypertuner tester le modèle, puis le tester finalement sur le test set.
```

```python
# Cross validation : on divise les données en k parties, et on entraine le modèle sur k-1 parties, puis on teste sur la dernière.
# On répète l'opération k fois, en changeant à chaque fois la partie de test.

from sklearn.model_selection import cross_val_score 

cross_val_score(KNeighborsClassifier(), X_train, y_train, cv=5, scoring='accuracy').mean() # métrique à ajuster selon le modèle
```

```python
# Validation curve : permet de regarder l'impact des hyperparamètres sur le modèle

from sklearn.model_selection import validation_curve

model = KNeighborsClassifier()
k = np.arange(1,15)

train_score, val_score = validation_curve(estimator=model, X=X_train, y=y_train, param_name='n_neighbors', param_range=k ) # on préciser l'hyperparamètre à tester
test_score, val_score_test = validation_curve(estimator=model, X=X_test, y=y_test, param_name='n_neighbors', param_range=k )

plt.plot(k, val_score.mean(axis=1), label='validation') # on affiche le score moyen en fonction des k
plt.plot(k, val_score_test.mean(axis=1), label='test')
plt.legend()
plt.show()
```

On peut ainsi repérer l'overfit en traçant les courbes pour le test et le train : si les scores de train augmentent mais ceux de test commencent à diminuer, on est probablement en train de sur-apprendre.

Avec GridSearch, on peut chercher les meilleurs hypersparamètres.

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'n_neighbors': np.arange(1,20), 'metric': ['euclidean', 'manhattan']} # on définit les hyperparamètres à tester

grid = GridSearchCV(KNeighborsClassifier(), param_grid=param_grid, cv=5) # on définit le modèle, les hyperparamètres, et le nombre de parties   

grid.fit(X_train, y_train)
```
